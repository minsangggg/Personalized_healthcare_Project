```python
import re

class IngredientETL:
    def __init__(self):
        # âœ… ê³ ê¸°ë¥˜ í‘œì¤€í™”
        self.meat_patterns = {
            r'ì†Œê³ ê¸°.*ì–‘ì§€|ì–‘ì§€ë¨¸ë¦¬|ì†Œê³ ê¸°ì–‘ì§€': 'ì†Œê³ ê¸°_ì–‘ì§€',
            r'ì†Œê³ ê¸°.*êµ­ê±°ë¦¬': 'ì†Œê³ ê¸°_êµ­ê±°ë¦¬',
            r'ì†Œê³ ê¸°.*ë‹¤ì§ìœ¡': 'ì†Œê³ ê¸°_ë‹¤ì§ìœ¡',
            r'ì‡ ê³ ê¸°': 'ì†Œê³ ê¸°_ì–‘ì§€',
            r'ë¼ì§€ê³ ê¸°.*ë‹¤ì§ìœ¡': 'ë¼ì§€ê³ ê¸°_ë‹¤ì§ìœ¡',
            r'ë¼ì§€ê³ ê¸°.*ì‚¼ê²¹ì‚´': 'ë¼ì§€ê³ ê¸°_ì‚¼ê²¹ì‚´',
            r'ë‹­ê°€ìŠ´ì‚´|ë‹­.*ê°€ìŠ´ì‚´': 'ë‹­ê³ ê¸°_ê°€ìŠ´ì‚´',
            r'ë‹­ë‹¤ë¦¬|ë‹­.*ë‹¤ë¦¬': 'ë‹­ê³ ê¸°_ë‹¤ë¦¬',
            r'ë² ì´ì»¨': 'ë² ì´ì»¨'
        }

        # âœ… ë¶ˆí•„ìš”í•œ ì„¤ëª… ì œê±°
        self.description_patterns = [
            (r'ê³„ë€.*í°ì|ë‹¬ê±€.*í°ì', 'ê³„ë€'),
            (r'ê³„ë€.*ë…¸ë¥¸ì|ë‹¬ê±€.*ë…¸ë¥¸ì', 'ê³„ë€'),
            (r'(ë¹¨ê°•|ë¹¨ê°„|ë…¸ë‘|ë…¸ë€|ì´ˆë¡|ì²­|í™|í™©)[ ]*íŒŒí”„ë¦¬ì¹´', 'íŒŒí”„ë¦¬ì¹´'),
            (r'(ë¹¨ê°•|ë¹¨ê°„|ë…¸ë‘|ë…¸ë€|ì´ˆë¡|ì²­|í™|í™©)[ ]*í”¼ë§', 'í”¼ë§'),
            (r'ë‹¤ì§„|ì¬|ê¹|ê°ˆì€|ê°€ë£¨|ê³±ê²Œ|êµµê²Œ|í¼ì§í•˜ê²Œ|ë¨¹ê¸°.*ì¢‹ê²Œ|í•œì….*í¬ê¸°', ''),
            (r'\([^)]*\)|\[[^\]]*\]', ''),  # ê´„í˜¸ ë‚´ìš© ì œê±°
        ]

        # âœ… ì¬ë£Œëª… í†µí•© ê·œì¹™ (í’ˆëª© ë‹¨ìœ„ë¡œ ë¬¶ê¸°)
        self.name_unify_patterns = {
            # ê¸°ë³¸ ì¬ë£Œ
            r'^ê°ì.*': 'ê°ì',
            r'^ê³ êµ¬ë§ˆ.*': 'ê³ êµ¬ë§ˆ',
            r'^ì–‘íŒŒ.*': 'ì–‘íŒŒ',
            r'^ëŒ€íŒŒ.*': 'ëŒ€íŒŒ',
            r'.*ë²„ì„¯.*': 'ë²„ì„¯',
            r'íœ˜í•‘í¬ë¦¼.*|ìƒí¬ë¦¼.*': 'íœ˜í•‘í¬ë¦¼',
            r'íŒŒí”„ë¦¬ì¹´.*': 'íŒŒí”„ë¦¬ì¹´',
            r'í”¼ë§.*': 'í”¼ë§',
            r'ì˜¤ì´.*': 'ì˜¤ì´',
            r'ë‹¬ê±€.*|ê³„ë€.*': 'ê³„ë€',
            r'ê¹€ì¹˜.*': 'ê¹€ì¹˜',
            r'^ë‹¹ê·¼.*': 'ë‹¹ê·¼',
            r'ë¸Œë¡œì½œë¦¬.*|ë¸Œë¡œì»¬ë¦¬.*': 'ë¸Œë¡œì½œë¦¬',
            r'ë‘ë¶€.*': 'ë‘ë¶€',
            r'.*ì¹˜ì¦ˆ.*': 'ì¹˜ì¦ˆ',
            r'ë¹„ì—”ë‚˜.*|ì†Œì„¸ì§€.*|ì†Œì‹œì§€.*|í–„.*|ìŠ¤íŒ¸.*': 'í–„/ì†Œì‹œì§€',
            r'ì°¸ì¹˜.*': 'ì°¸ì¹˜',
            r'ì–‘ë°°ì¶”.*': 'ì–‘ë°°ì¶”',
            r'ìŒ€.*|ë°¥.*': 'ìŒ€/ë°¥',
        }

        # âœ… í¬ê¸°/ìˆ˜ì‹ì–´ ì œê±°ìš© ë‹¨ì–´
        self.redundant_modifiers = [
            'ì‘ì€', 'í°', 'ì¤‘ê°„', 'ê±°', 'ê²ƒ', 'ì‚¬ì´ì¦ˆ', 'ì†Œ', 'ëŒ€', 'ì¤‘', 
            'ë¨¹ì„ë§Œí¼', 'íŒŒë¦¬', 'ì±„', 'íŒŒ', 'ì¡°ê°', 'ì¡°ê·¸ë§Œ', 'ì¡°ê¸ˆ', 'í•œì…'
        ]

    def remove_description(self, ingredient: str) -> str:
        """ ë¶ˆí•„ìš”í•œ ë‹¨ì–´ ì œê±° """
        result = ingredient
        for pattern, replacement in self.description_patterns:
            result = re.sub(pattern, replacement, result)
        return ' '.join(result.split()).strip()

    def normalize_meat(self, ingredient: str) -> str:
        """ ê³ ê¸°ë¥˜ í‘œì¤€í™” """
        for pattern, standard in self.meat_patterns.items():
            if re.search(pattern, ingredient):
                return standard
        return ingredient.strip()

    def unify_name(self, ingredient: str) -> str:
        """ ì¬ë£Œëª… í†µí•© (í¬ê¸°/ìˆ˜ì‹ì–´ ì œê±° + í’ˆëª© ë‹¨ìœ„ í†µì¼) """
        name = ingredient

        # 1ï¸âƒ£ í¬ê¸°Â·í˜•ìš©ì‚¬ ì œê±°
        for word in self.redundant_modifiers:
            name = re.sub(word, '', name)

        # 2ï¸âƒ£ í’ˆëª©ëª… í†µí•©
        for pattern, unified in self.name_unify_patterns.items():
            if re.search(pattern, name):
                name = unified
                break

        # 3ï¸âƒ£ ê³µë°± ì •ë¦¬
        return name.strip()
```
