```python
import re
class IngredientETL:
    def __init__(self):
        # ----------------------------------
        # 🥩 1️⃣ 고기류 표준화
        # ----------------------------------
        self.meat_patterns = {
            r'소고기.*양지|양지머리|소고기양지': '소고기_양지',
            r'소고기.*국거리|국거리용 소고기': '소고기_국거리',
            r'소고기.*갈비찜': '소고기_갈비찜',
            r'소고기.*등심': '소고기_등심',
            r'쇠고기': '소고기_양지',
            r'돼지고기.*다짐육|간 돼지고기|간돼지고기|돈다짐육|돼지고기간것|돼지고기것': '돼지고기_다짐육',
            r'돼지고기.*삼겹살|대패삼겹살|대패돼지고기|대패 삼겹살': '돼지고기_삼겹살',
            r'돼지고기.*등심': '돼지고기_등심',
            r'돼지고기.*목살': '돼지고기_목살',
            r'돼지고기.*안심': '돼지고기_안심',
            r'돼지고기.*사태': '돼지고기_사태',
            r'돼지고기.*찌개': '돼지고기_찌개',
            r'돼지고기.*오겹살': '돼지고기_오겹살',
            r'돼지고기.*불고기': '돼지고기_불고기',
            r'돼지고기.*등뼈|감자탕용 돼지등뼈': '돼지고기_등뼈',
            r'돼지갈비': '돼지고기_갈비',
            r'닭.*가슴살': '닭고기_가슴살',
            r'닭.*다리': '닭고기_다리',
            r'닭.*날개': '닭고기_날개',
            r'닭.*닭봉': '닭고기_닭봉',
            r'닭.*모래집': '닭고기_모래집',
            r'닭.*안심': '닭고기_안심',
            r'닭.*볶음탕|볶음용': '닭고기_닭볶음탕',
            r'닭 한마리|닭고기': '닭고기',
            r'베이컨': '베이컨',
        }

        # ----------------------------------
        # ✂️ 2️⃣ 불필요한 단어·형용사 제거
        # ----------------------------------
        self.description_patterns = [
            (r'계란.*흰자|달걀.*흰자', '계란'),
            (r'계란.*노른자|달걀.*노른자', '계란'),
            (r'(빨강|빨간|노랑|노란|초록|청|홍|황)[ ]*파프리카', '파프리카'),
            (r'(빨강|빨간|노랑|노란|초록|청|홍|황)[ ]*피망', '피망'),
            (r'다진 마늘', '다진마늘'),
            (r'다진 생강', '다진생강'),
            (r'간 마늘', '다진마늘'),
            (r'간 생강', '다진생강'),
            (r'간 양파', '다진양파'),
            (r'각종.*채소|각종자투리.*야채|남은야채들|냉장고속.*채소', '채소'),
            (r'채소.*|야채.*', '채소'),
            (r'구운[ ]*김|김밥[ ]*김|도시락[ ]*김', '김'),
            (r'묵은지|김장김치|배추김치.*', '김치'),
            (r'공기밥|밥', '쌀/밥'),
            (r'냉동.*새우|꽃새우', '새우'),
            (r'냉동.*만두|군만두', '만두'),
            (r'건표고.*|표고버섯.*', '표고버섯'),
            (r'건미역|미역.*', '미역'),
            (r'건멸치|다시멸치|국멸치', '멸치'),
            (r'건크랜베리|건 크랜베리', '건크랜베리'),
            (r'고구마.*', '고구마'),
            (r'감자(?!탕)', '감자'),  # 감자탕 제외
            (r'대파.*', '파'),
            (r'파 흰부분|대파 흰부분|파 초록부분', '파'),
            (r'파뿌리|대파뿌리', '파뿌리'),
            (r'비엔나.*|소세지.*|소시지.*|햄.*|스팸.*', '햄/소시지'),
        ]

        # ----------------------------------
        # 🧩 3️⃣ 품목 통합 (마지막 레이어)
        # ----------------------------------
        self.name_unify_patterns = {
            r'^감자.*': '감자',
            r'^고구마.*': '고구마',
            r'^양파.*': '양파',
            r'^대파.*|파$': '파',
            r'.*버섯.*': '버섯',
            r'휘핑크림.*|생크림.*': '휘핑크림',
            r'파프리카.*': '파프리카',
            r'피망.*': '피망',
            r'오이.*': '오이',
            r'달걀.*|계란.*': '계란',
            r'김치.*': '김치',
            r'^당근.*': '당근',
            r'브로콜리.*|브로컬리.*': '브로콜리',
            r'두부.*': '두부',
            r'.*치즈.*': '치즈',
            r'참치.*': '참치',
            r'양배추.*': '양배추',
            r'쌀.*|밥.*': '쌀/밥',
            r'채소.*|야채.*': '채소',
        }
        
        self.redundant_modifiers = [
        "다진", "썬", "잘게", "채썬", "간", "볶은", "삶은", "데친", 
        "익힌", "말린", "껍질벗긴", "씨뺀", "통째로", "조각낸", 
        "조각", "슬라이스", "얇게썬", "잘게썬", "큰", "작은", "적당히", "조금"
        ]

    # -------------------------
    # 이하 함수들 순서 유지
    # -------------------------
    def remove_description(self, ingredient: str) -> str:
        """ 불필요한 단어 제거 """
        result = ingredient
        for pattern, replacement in self.description_patterns:
            result = re.sub(pattern, replacement, result)
        return ' '.join(result.split()).strip()


    def normalize_meat(self, ingredient: str) -> str:
        """ 고기류 표준화 """
        for pattern, standard in self.meat_patterns.items():
            if re.search(pattern, ingredient):
                return standard
        return ingredient.strip()


    def unify_name(self, ingredient: str) -> str:
        """ 재료명 통합 (크기/수식어 제거 + 품목 단위 통일) """
        name = ingredient

        # 1️⃣ 크기·형용사 제거
        for word in self.redundant_modifiers:
            name = re.sub(word, '', name)

        # 2️⃣ 품목명 통합
        for pattern, unified in self.name_unify_patterns.items():
            if re.search(pattern, name):
                name = unified
                break

        # 3️⃣ 공백 정리
        return name.strip()
```
